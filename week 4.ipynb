```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the GDP data
nepal_economy = pd.read_csv('Data/nepal-economy.csv')
nepal_economy['no_of_people_in_foreign_employment'] = nepal_economy['no_of_people_in_foreign_employment'].apply(lambda n :n/1000)

display(nepal_economy.head())
plt.figure(figsize=(5, 3))
plt.scatter(nepal_economy['GDP'],
    nepal_economy['no_of_people_in_foreign_employment'], 
    color='purple', 
    edgecolor='black'
)
# fit a lineat regression model 
x= nepal_economy['GDP']
y= nepal_economy['no_of_people_in_foreign_employment']
m,c = np.polyfit(x,y,1)

plt.plot(x,m*x+c, color= 'red', linewidth=1, label='GDP-FE-trend_line')
#add titles and labels
plt.xticks(rotation=45)
plt.title('GDP vs No Of people in foreign employment ')
plt.xlabel('GDP(USD Billion)')
plt.ylabel('People in Foreign Employment (K)')
plt.grid(True)
# Display the plot
plt.show()
```

## Bubble Graph

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the GDP data
nepal_economy = pd.read_csv('Data/nepal-economy.csv')
nepal_economy['no_of_people_in_foreign_employment'] = nepal_economy['no_of_people_in_foreign_employment'].apply(lambda n: n / 1000)

# Display the first few rows of the DataFrame
display(nepal_economy.head())

# Set up the figure size
plt.figure(figsize=(10, 6))

# Create a bubble chart
plt.scatter(
    nepal_economy['GDP'],
    nepal_economy['no_of_people_in_foreign_employment'],
    s=nepal_economy['no_of_people_in_foreign_employment'] * 10,  # Bubble size (scaled)
    color='cyan',
    edgecolor='black',
    alpha=0.6
)

# Fit a linear regression model
x = nepal_economy['GDP']
y = nepal_economy['no_of_people_in_foreign_employment']
m, c = np.polyfit(x, y, 1)

# Plot the trend line
plt.plot(x, m * x + c, color='red', linewidth=1, label='GDP-FE-trend_line')

# Add titles and labels
plt.xticks(rotation=45)
plt.title('Bubble Chart: GDP vs No Of People in Foreign Employment')
plt.xlabel('GDP (USD Billion)')
plt.ylabel('People in Foreign Employment (K)')
plt.grid(True)

# Display the plot
plt.legend()
plt.show()
```

## Distributions
- Histogram: Histogram is used to describe numecical data which have grouped into
frequency distribution
- Frequency Polygon: Frequency Polygon can be constructed simply by joining the
upper mid point of histogram
- Cumulative Frequency Distribution(ogive): Cumulative frequency distribtion
facilitages to see how may observations lie above or below certain values and is also
graphicial presentation of frequency distribution.
- Box Plot: It shows the Five Point summary (Min,Q1, Median, Q3, Max) and outliers

### Histogram

```python
students_data = pd.DataFrame({
    'Marks':['0-20','20-40','40-60','60-80','80-100'],
    'no_of_students':[10,22,35,28,5]
})
students_data['mid_points'] = students_data['Marks'].apply(lambda m:( int( m.split('-')[0]) + int( m.split('-')[1]))/2)
lower_bounds= students_data[ 'Marks'].apply(lambda m: int(m.split('-')[0]) ).min()
upper_bounds =  students_data['Marks'].apply(lambda m: int(m.split('-')[1])).max()

bin_edges = list(range(lower_bounds, upper_bounds + 1, 20))
students_data['bins'] = pd.cut(x=students_data['no_of_students'], bins=bin_edges)
print(bin_edges)
display(students_data.head())
plt.figure(figsize=(10,5))

# use mid points as the x-values for the histogram and the frequency y-values
plt.hist(students_data['mid_points'], bins=bin_edges, weights=students_data['no_of_students'], edgecolor='black', alpha=0.7, label='Marks')
#add labels and titles
plt.xlabel('marks range (Mid-Points)')
plt.ylabel('number of students')
plt.title('histogram with frequesncy curve')

# Plot frequency polygon
plt.plot(students_data['mid_points'], students_data['no_of_students'], 
         marker='o', color='orange', linestyle='-', label='Frequency Polygon')

plt.legend()
plt.tight_layout()
plt.show()

```

### Frequency curve 


```python
import pandas as pd
import matplotlib.pyplot as plt

# Original data
students_data = pd.DataFrame({
    'Marks': ['0-20', '20-40', '40-60', '60-80', '80-100'],
    'no_of_students': [10, 22, 35, 28, 5]
})

# Calculate mid points
students_data['mid_points'] = students_data['Marks'].apply(lambda m: (int(m.split('-')[0]) + int(m.split('-')[1])) / 2)

# Calculate lower and upper bounds
lower_bounds = students_data['Marks'].apply(lambda m: int(m.split('-')[0])).min()
upper_bounds = students_data['Marks'].apply(lambda m: int(m.split('-')[1])).max()

# Create bin edges
bin_edges = list(range(lower_bounds, upper_bounds + 1, 20))

# Calculate cumulative frequency
students_data['cumulative_frequency'] = students_data['no_of_students'].cumsum()

# Plotting
plt.figure(figsize=(19, 6))

# Plot frequency curve
plt.subplot(1, 2, 1)
plt.plot(students_data['mid_points'], students_data['no_of_students'], marker='o', linestyle='-', color='b', label='Frequency Curve')
plt.title('Frequency Curve')
plt.xlabel('Marks Range (Mid-Points)')
plt.ylabel('Number of Students')
plt.xticks(students_data['mid_points'])
plt.grid()
plt.legend()



plt.tight_layout()
plt.show()
```

### Stacked Bar Graph

```python
import json
import matplotlib.pyplot as plt
import pandas as pd


# Define the ethnic distribution data
ethnic_distribution = {
    "Newar": {"male": 3000, "female": 3200},
    "Tharu": {"male": 2500, "female": 2700},
    "Chettri": {"male": 4000, "female": 3800},
    "Brahmin": {"male": 1500, "female": 1600},
    "Tamang": {"male": 2000, "female": 2100}
}

# Specify the file name
file_name = 'Data/nepal-population-ethinic-group.json'

# Write the data to a JSON file
with open(file_name, 'w') as json_file:
    json.dump(ethnic_distribution, json_file, indent=4)

print(f"Data successfully written to {file_name}")

# Load the ethnic distribution data
with open('Data/nepal-population-ethinic-group.json', 'r') as file:
    ethnic_dist = json.load(file)

# Data Preparation
# Extracting ethnic groups, male and female populations
labels = list(ethnic_dist.keys())
male_sizes = [ethnic_dist[group]['male'] for group in labels]
female_sizes = [ethnic_dist[group]['female'] for group in labels]

# Create a DataFrame for better handling (optional)
data = pd.DataFrame({
    'Ethnic Group': labels,
    'Male Population': male_sizes,
    'Female Population': female_sizes
})

# Create the stacked bar graph
plt.figure(figsize=(10, 6))

# Plotting the stacked bar graph
plt.bar(data['Ethnic Group'], data['Male Population'], color='blue', label='Male')
plt.bar(data['Ethnic Group'], data['Female Population'], bottom=data['Male Population'], color='pink', label='Female')

# Add labels and title
plt.xlabel('Ethnic Groups')
plt.ylabel('Population')
plt.title('Ethnic Distribution in Nepal by Gender')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
