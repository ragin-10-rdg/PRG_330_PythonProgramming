## Basic Indexing

```python
print(students["name"])
#Single column as a Series
new_std_data = students[["id", "name", "mark"]].copy()
print(new_std_data) # Multiple columns as a DataFrame


# # # Slicing rowS


first_10_rows = students[:10].copy()
print(first_10_rows) # only first row (dataframe)
last_five_rows = students[30:].copy()
print(last_five_rows)
# Rows from index 30 to last index (datafrane)
```

## Label-based Indexing (.loc[])

```python
import pandas as pd

# Read the CSV file
students = pd.read_csv('Data/student.csv')

# Select the first row
first_row = students.loc[0]
print("First Row:")
print(first_row)

# Select the first two rows
first_two_rows = students.loc[0:1]
print("\nFirst Two Rows:")
print(first_two_rows)

# Select specific rows and columns (rows 0 to 2, columns from 'name' to 'mark')
first_two_stds_with_name_to_marks = students.loc[0:2, 'name':'mark']
print("\nFirst Two Students with Name to Marks:")
print(first_two_stds_with_name_to_marks)

# Select specific rows (0 and 1) and specific columns ('name' and 'mark')
selected_rows_and_columns = students.loc[[5, 2], ['name', 'mark']]
print("\nSelected Rows 0 and 1 with Columns 'name' and 'mark':")
print(selected_rows_and_columns)

# Select specific rows (5 and 20) and specific columns ('name' and 'gender')
first_and_second_row_with_name_and_gender = students.loc[[5, 20], ['name', 'gender']]
print("Rows 5 and 20 with columns 'name' and 'gender':")
print(first_and_second_row_with_name_and_gender)
```

## Integer-Label-based Indexing (.iloc[])

```python
#Selecting rows based on index positions
print(students.iloc[1])
print(students.iloc[0:3])
#Selecting specific rows and columns
print(students.iloc[0:2, 0:2])
print(students.iloc[[0, 2], [0, 2]])
```

## Single Value Access (.at[], .iat[])
- .at[] is used for accessing a single scalar value by label.
- .iat[] is for accessing by index position.

```python
import pandas as pd
students = pd.read_csv('Data/student.csv')

# Using .at[] to access single value by label
print(students.at[1, 'name'])                # Value in row 2, column 'name'
# Using .iat[] to access single value by position
print(students.iat[5, 1])                 # Value in row 6, column 1
```
